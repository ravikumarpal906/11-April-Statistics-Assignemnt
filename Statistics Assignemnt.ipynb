**3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.**

```python
class DiscreteRandomVariable:
  def __init__(self,outcomes,probabilities):
    self.outcomes = outcomes
    self.probabilities = probabilities

  def expected_value(self):
    return sum(outcome*probability for outcome,probability in zip(self.outcomes,self.probabilities))

  def variance(self):
    mean = self.expected_value()
    return sum((outcome-mean)**2*probability for outcome,probability in zip(self.outcomes,self.probabilities))

values = [0,1,2,3]
probabilities = [0.25,0.35,0.25,0.15]

random_variable = DiscreteRandomVariable(values,probabilities)

print("Expected value:",random_variable.expected_value())
print("Variance:",random_variable.variance())
```

**5. Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.**

```python
import numpy as np

def generate_samples(distribution, params, n):
  """
  Generates random samples from a given probability distribution.

  Args:
    distribution: A string representing the probability distribution (e.g., "binomial", "poisson").
    params: A list of parameters for the distribution.
    n: The number of samples to generate.

  Returns:
    A numpy array of random samples.
  """

  if distribution == "binomial":
    return np.random.binomial(params[0], params[1], n)
  elif distribution == "poisson":
    return np.random.poisson(params[0], n)
  else:
    raise ValueError("Invalid distribution:", distribution)

def calculate_mean_variance(samples):
  """
  Calculates the mean and variance of a given set of samples.

  Args:
    samples: A numpy array of random samples.

  Returns:
    A tuple containing the mean and variance of the samples.
  """

  mean = np.mean(samples)
  variance = np.var(samples)
  return mean, variance

# Generate random samples from a binomial distribution
samples = generate_samples("binomial", [10, 0.5], 1000)

# Calculate the mean and variance
mean, variance = calculate_mean_variance(samples)

# Print the results
print("Mean:", mean)
print("Variance:", variance)
```

```python
def skewness(data):
  """
  Calculates the skewness of a given dataset.

  Args:
    data: A numpy array or pandas Series containing the data.

  Returns:
    The skewness of the data.
  """

  mean = np.mean(data)
  std_dev = np.std(data)
  return (3 * (mean - np.median(data))) / std_dev

# Calculate the skewness of total_bill and tip columns
total_bill_skewness = skewness(tips['total_bill'])
tip_skewness = skewness(tips['tip'])

# Print the results
print("Total bill skewness:", total_bill_skewness)
print("Tip skewness:", tip_skewness)
```

(iii) Write a function that calculates the covariance between two columns.

```python
def covariance(data1, data2):
  """
  Calculates the covariance between two datasets.

  Args:
    data1: A numpy array or pandas Series containing the first dataset.
    data2: A numpy array or pandas Series containing the second dataset.

  Returns:
    The covariance between the two datasets.
  """

  mean1 = np.mean(data1)
  mean2 = np.mean(data2)
  return np.sum((data1 - mean1) * (data2 - mean2)) / len(data1)

# Calculate the covariance between total_bill and tip columns
covariance_total_bill_tip = covariance(tips['total_bill'], tips['tip'])

# Print the result
print("Covariance between total_bill and tip:", covariance_total_bill_tip)
```

(iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns.

```python
def correlation_coefficient(data1, data2):
  """
  Calculates the Pearson correlation coefficient between two datasets.

  Args:
    data1: A numpy array or pandas Series containing the first dataset.
    data2: A numpy array or pandas Series containing the second dataset.

  Returns:
    The Pearson correlation coefficient between the two datasets.
  """

  covariance_total_bill_tip = covariance(data1, data2)
  std_dev_total_bill = np.std(data1)
  std_dev_tip = np.std(data2)
  return covariance_total_bill_tip / (std_dev_total_bill * std_dev_tip)

# Calculate the Pearson correlation coefficient between total_bill and tip columns
correlation_coefficient_total_bill_tip = correlation_coefficient(tips['total_bill'], tips['tip'])

# Print the result
print("Pearson correlation coefficient between total_bill and tip:", correlation_coefficient_total_bill_tip)
```

**8. Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.**

```python
import math

def normal_pdf(x, mu, sigma):
  """
  Calculates the probability density function (PDF) of a continuous random variable for a given normal distribution.

  Args:
    x: The value at which to evaluate the PDF.
    mu: The mean of the normal distribution.
    sigma: The standard deviation of the normal distribution.

  Returns:
    The probability density function (PDF) of x.
  """

  return (1 / (sigma * math.sqrt(2 * math.pi))) * math.exp(-((x - mu)**2) / (2 * sigma**2))
```

**9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.**

```python
import math

def exponential_cdf(x, lambda_):
  """
  Calculates the cumulative distribution function (CDF) of an exponential distribution.

  Args:
    x: The value at which to evaluate the CDF.
    lambda_: The rate parameter of the exponential distribution.

  Returns:
    The cumulative distribution function (CDF) of x.
  """

  return 1 - math.exp(-lambda_ * x)

# Example usage
lambda_ = 0.5
x = 1

cdf = exponential_cdf(x, lambda_)

print("The CDF of", x, "for an exponential distribution with lambda =", lambda_, "is", cdf)
```

**10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.**

```python
def poisson_pmf(k, lambda_):
  """
  Calculates the probability mass function (PMF) of a Poisson distribution.

  Args:
    k: The number of successes.
    lambda_: The rate parameter of the Poisson distribution.

  Returns:
    The probability mass function (PMF) of k.
  """

  return (lambda_**k * math.exp(-lambda_)) / math.factorial(k)

# Example usage
k = 5
lambda_ = 2

pmf = poisson_pmf(k, lambda_)

print("The PMF of", k, "for a Poisson distribution with lambda =", lambda_, "is", pmf)
```

**11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare.**

```python
# To generate the data use the following command:

import numpy as np
#50 purchases out of 1000 visitors
old_layout = np.array([1]*50+ [0]*950)
#70 purchases out of 1000 visitors
new_layout = np.array([1]*70+ [0] *930)
# Apply z-test to find which layout is successful.

```

```python
import scipy.stats as stats

# Calculate the sample proportions
old_proportion = np.mean(old_layout)
new_proportion = np.mean(new_layout)

# Calculate the standard errors
old_se = np.sqrt(old_proportion * (1 - old_proportion) / len(old_layout))
new_se = np.sqrt(new_proportion * (1 - new_proportion) / len(new_layout))

# Calculate the z-statistic
z = (new_proportion - old_proportion) / np.sqrt(old_se**2 + new_se**2)

# Calculate the p-value
p_value = stats.norm.cdf(z)

# Compare the p-value to the significance level (alpha)
alpha = 0.05
if p_value < alpha:
  print("Reject the null hypothesis. The new layout is more successful.")
else:
  print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the new layout is more successful.")
```

**15. A company is testing two different website layouts to see which one leads to higher click-through rates. Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value**

```python
# Use the following data:

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30,35,36,37]
layout_b_clicks = [40,41,38,42,39,44,43,41,45,47]

import scipy.stats as stats

def ab_test(control_data, treatment_data):
  """
  Performs an A/B test analysis.

  Args:
    control_data: A list of data points for the control group.
    treatment_data: A list of data points for the treatment group.

  Returns:
    A tuple containing the t-statistic, degrees of freedom, and p-value.
  """

  # Calculate the means and variances
  control_mean = np.mean(control_data)
  control_var = np.var(control_data, ddof=1)
  treatment_mean = np.mean(treatment_data)
  treatment_var = np.var(treatment_data, ddof=1)

  # Calculate the pooled variance
  pooled_var = ((len(control_data) - 1) * control_var + (len(treatment_data) - 1) * treatment_var) / (len(control_data) + len(treatment_data) - 2)

  # Calculate the standard error of the difference between means
  std_err = np.sqrt(pooled_var * (1 / len(control_data) + 1 / len(treatment_data)))

  # Calculate the t-statistic
  t_stat = (treatment_mean - control_mean) / std_err

  # Calculate the degrees of freedom
  df = len(control_data) + len(treatment_data) - 2

  # Calculate the p-value
  p_value = stats.t.cdf(-abs(t_stat), df) * 2

  # Return the results
  return t_stat, df, p_value

# Perform the A/B test
t_stat, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)

# Print the results
print("t-statistic:", t_stat)
print("degrees of freedom:", df)
print("p-value:", p_value)
```

**21. A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences.**

```python
import numpy as np
from scipy.stats import chi2_contingency

# Null hypothesis: There is no association between age groups and voter preferences.
# Alternate hypothesis: There is a significant association between age groups and voter preferences.

# Set the random seed for reproducibility
np.random.seed(0)

# Generate the data
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500, p=[0.3, 0.4, 0.3])
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500, p=[0.5, 0.5])

# Create a contingency table
data = {'Age Group': age_groups, 'Voter Preference': voter_preferences}
df = pd.DataFrame(data)
contingency_table = pd.crosstab(df['Age Group'], df['Voter Preference'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p}")
print("Expected Frequencies:")
print(expected)

if p < 0.05:
  print("Reject the null hypothesis. There is a significant association between age groups and voter preferences.")
else:
  print("Fail to reject the null hypothesis. There is not enough evidence to conclude that there is a significant association between age groups and voter preferences.")
```

The survey data is summarized in a contingency table. Conduct a Chi- Square test to determine if there is a significant relationship between product satisfaction levels and customer regions.**

```python
# Sample data: python
#Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)
data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])

# Null hypothesis: There is no relationship between product satisfaction levels and customer regions.
# Alternate hypothesis: There is a significant relationship between product satisfaction levels and customer regions.

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(data)

data = pd.DataFrame(data, columns=['East', 'West', 'North', 'South'], index=['Satisfied', 'Neutral', 'Dissatisfied'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi-Square Statistic: {chi2}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p}")
print("Expected Frequencies:")
print(expected)

if p < 0.05:
  print("Reject the null hypothesis. There is a significant relationship between product satisfaction levels and customer regions.")
else:
  print('fail to reject the null hypothesis')
```

**23. A company implemented an employee training program to improve job performance (Effective, Neutral, Ineffective). After the training, they collected data from a sample of employees and classified them based on their job performance before and after the training. Perform a Chi-Square test to determine if there is a significant difference between job performance levels before and after the training.**

```python
# Sample data: python
# Sample data: Job performance levels before (rows) and after (columns) training
data = np.array([[50, 30, 20], [30, 40, 30], [20, 30, 40]])

data = pd.DataFrame(data, columns=['Effective', 'Neutral', 'Ineffective'], index=['Before', 'After', 'Difference'])

chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi-Square Statistic: {chi2}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p}")
print("Expected Frequencies:")
print(expected)

if p < 0.05:
  print("Reject the null hypothesis. There is a significant difference between job performance levels before and after the training.")
else:
  print("Fail to reject the null hypothesis. There is not enough evidence to conclude that there is a significant difference between job performance levels before and after the training.")
```
